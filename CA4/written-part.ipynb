#### *How to find the best regression using gradient descent method?*
1. Choosing a cost function (Here we use MSE)
2. Initializing the weights. Starting with random or zero values.
3. Calculating the gradient. It involves computing the partial derivatives of the cost function with respect to each parameter. The gradient indicates the direction of the steepest ascent, and the opposite direction is the direction of the steepest descent.
4. Updating the weights by subtracting the product of the learning rate and the gradient. This moves the coefficients towards the minimum of the cost function.
Now we Iterate until convergence. It means to repeat steps 3 and 4 until the change in the coefficients becomes smaller than a predefined threshold or a maximum number of iterations is reached. This ensures that the algorithm converges to a local minimum of the cost function.

#### *Other method to find the best regression*
Adam optimizer: It uses estimates of the first and second moments of the gradients to adjust the learning rate and maintains an exponential moving average of past gradients and squared gradients. This adaptivity can lead to faster convergence and better performance on complex problems.

## Q5. Support Vector Machine
#### *Which points are called support vectors?*
They are the data points that lie closest to the decision boundary (hyperplane) and influence its position. These points are more important in the classification process because they determine the optimal separation between different classes.
#### *What datasets are SVM algorithms not suitable for?*
datasets with a large number of features, especially when the number of features is significantly larger than the number of samples. This is because SVMs can become computationally expensive and may suffer from overfitting in such cases. Additionally, SVMs might not perform well on noisy datasets or datasets with many overlapping classes, as they rely on finding a clear margin of separation.
#### *kernels*
They are functions that calculate the dot product of two vectors in a higher dimensional space, allowing SVMs to handle non-linear boundaries between classes. Kernel functions project the original input data into a higher dimensional feature space, where it becomes linearly separable. Common kernel functions include linear, polynomial, RBF (for radial basis function), and sigmoid.
#### *What are the differences between soft and hard SVM?*
They're differ in how they handle misclassifications:
Hard SVM aims to find a hyperplane that linearly separates all data points with the largest possible margin. However, in real-world datasets, perfect separation may not always be possible. In such cases, hard SVM might fail to generalize well or even fail to find a solution.
Soft SVM tries to solve the limitations of hard SVM, soft SVM introduces a soft margin that allows for some misclassifications. This is achieved by introducing a regularization parameter (C) that controls the trade-off between maximizing the margin and minimizing misclassifications.
