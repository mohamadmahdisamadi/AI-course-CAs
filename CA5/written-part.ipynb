#### 3. Using backward propagation and gradient descent, update $ w_5 $

Let $ \mathcal{L} $ represent the loss function.

Let $ \alpha $ represent the learning rate.

We're asked to update only $ w_5 $ using backward propagation algorithm. Our goal is to reduce the error by changing only $ w_5 $. In order to achive such goal we'll calculate the gradient of $ \mathcal{L} $ with respect to $ w_5 $ using chain rule. Here's the procedure:

$ \frac{\partial \mathcal{L}}{\partial w_5} = \frac{\partial \mathcal{L}}{\partial (\text{output }o_1)} \times \frac{\partial (\text{output }o_1)}{\partial (\text{input }o_1)} \times \frac{\partial (\text{input }o_1)}{\partial w_5} $

Let's calculate them one by one:

$ \mathcal{L} = \frac{1}{2}(\text{predicted label}_1 - \text{actual label}_1)^2 + (\text{predicted label}_2 - \text{actual label}_2)^2 $

$ \frac{\partial \mathcal{L}}{\partial (\text{output }o_1)} = 2 \times \frac{1}{2}(\text{predicted label}_1 - \text{actual label}_1)^{2 - 1} \times -1 + 0 $

$ \frac{\partial \mathcal{L}}{\partial (\text{output }o_1)} = -(\text{predicted label}_1 - \text{actual label}_1) = 0.6575  $


$ \frac{\partial (\text{output }o_1)}{\partial (\text{input }o_1)} = \frac{\partial (\text{sigmoid}(z))}{\partial (z)} = \frac{\partial(\frac{1}{1 + e^{-z}})}{\partial (z)} = \frac{\partial (1 + e^{-z})^{-1}}{\partial (z)} = -1 \times (1 + e^{-z})^{-2} \times (0 - e^{-z}) = \frac{e^{-z}}{(1 + e^{-z})^{2}} = 0.097  $

$ \frac{\partial (\text{input }o_1)}{\partial w_5} = \frac{\partial (w_5 \times \text{output }h_1 + w_7 \times \text{output }h_2 + b_3)}{\partial w_5} = \text{output }h_1 + 0 + 0 = 0.57 $

$ \frac{\partial \mathcal{L}}{\partial w_5}  = 0.6575 \times 0.097 \times 0.57 = 0.0363 \approx 0.04 $

$ \text{step size} = \text{Learning Rate} \times \text{Gradient} = 0.5 \times 0.04 = 0.02 $
